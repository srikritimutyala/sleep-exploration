```python
import numpy as np
import os

sleep_vectors_path = '/home/jupyter/rw-migration-local/data/sleep_vectors/'

sleep_vectors = {}
failed = []

for filename in os.listdir(sleep_vectors_path):
    if filename.endswith('.npy'):
        participant_id = filename.replace('.npy', '')
        filepath = os.path.join(sleep_vectors_path, filename)
        try:
            sleep_vectors[participant_id] = np.load(filepath, allow_pickle=True)
        except Exception as e:
            failed.append(participant_id)

print(f"Loaded: {len(sleep_vectors)} vectors")
print(f"Failed: {len(failed)} vectors")
```
